```python
import os
import ast
import shutil
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
login_url = "https://courses.edovo.com/login/check"
content_url = "https://courses.edovo.com/content/lessons/6986"
```

```python
payload='email=XXXXXXX%40XXXXXXX&password=XXXXXXX'
```

```python
headers = {
  'cookie': 'XXXXXXX',
  'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Mobile Safari/537.36',
  'Content-Type': 'application/x-www-form-urlencoded'
}
```

```python
def get_soup(url):
  response = requests.request("GET", url, headers=headers, data=payload)
  soup = BeautifulSoup(response.text, 'html.parser')
  return soup
```

# Get lessons

```python
soup = get_soup(content_url)

lessons = soup.find_all('li', attrs={'data-lesson': True})

lessons_info = []

for lesson in lessons:
  lesson_dic = {}

  title = lesson.find('div', attrs={'class': "list-lessons-text"}).text
  info = f"https://courses.edovo.com{lesson.find('a', attrs={'data-link': 'aw'}).get('href')}"
  pages = f"https://courses.edovo.com{lesson.find('a', attrs={'type': 'button'}).get('href')}"

  lesson_dic['title'] = title.strip()
  lesson_dic['info_link'] = info
  lesson_dic['pages_link'] = pages

  lessons_info.append(lesson_dic)
```

```python
df_lessons_info = pd.DataFrame.from_dict(lessons_info)
df_lessons_info
```

```python
df_lessons_info.to_csv("df_lessons_info.csv", index=False)
```

# Get lessons' info

```python
def get_description(url):
  info = {}

  description = get_soup(url)
  lesson_name = description.find('input', attrs={'placeholder': 'Lesson name'}).get('value')
  lesson_description = description.find('textarea', attrs={'placeholder': 'Lesson description'}).text

  info['lesson_name'] = lesson_name
  info['lesson_description'] = lesson_description

  return info
```

```python
lessons_info_list = []

for info_link in df_lessons_info.info_link.values:
  lesson_info = get_description(info_link)
  lessons_info_list.append(lesson_info)
```

```python
df_lessons_info_list = pd.DataFrame.from_dict(lessons_info_list)
df_lessons_info_list
```

```python
df_lessons_info_list.to_csv("df_lessons_info_list.csv", index=False)
```

# Get lessons' pages' links

```python
def get_pages(url, lesson):
  body = get_soup(url)

  try:
    body_script = body.find_all('script', attrs={'type': 'text/javascript'})[1].text

    for blurb in body_script.split("\n"):
      if "window.editor.lesson" in blurb:
        pages = blurb

    pages = ast.literal_eval(pages.split("=")[-1][:-1])['pages']
    pages_list = []

    for page in pages:
      page_dict = {}
      page_name = page['preview']
      page_id = page['id']

      page_dict['page_lesson'] = lesson
      page_dict['page_name'] = page_name
      page_dict['page_id'] = page_id

      pages_list.append(page_dict)

    return pages_list

  except:
    print(f"Issue with {lesson}")
    return []
```

```python
all_pages = []

for row in df_lessons_info.iterrows():
  lesson = row[1]['title'].split(":")[0]
  url = row[1]['pages_link']

  result = get_pages(url, lesson)
  all_pages.extend(result)
```

```python
df_all_pages = pd.DataFrame.from_dict(all_pages)
df_all_pages['page_number'] = df_all_pages.groupby('page_lesson').cumcount() + 1
df_all_pages['page_number'] = 'Page ' + df_all_pages['page_number'].astype(str)
df_all_pages
```

```python
df_all_pages.to_csv("df_all_pages.csv", index=False)
```

# Get each page

```python
def save_html(url_id, lesson, page):
  try:
    lesson_folder_name = lesson.lower().replace(" ", "_")
    page_folder_name = page.lower().replace(" ", "_")
    os.makedirs(f"raw_html/{lesson_folder_name}", exist_ok=True)

    page_url = f"https://courses.edovo.com/lesson/get_page_items?pageId={url_id}"
    page_soup = get_soup(page_url)
    html_str = page_soup.prettify()

    with open(f'raw_html/{lesson_folder_name}/{page_folder_name}.html', 'w', encoding='utf-8') as file:
      file.write(html_str)

    print(f"{lesson} - {page} completed.")
  except:
    print(f"ISSUE with {lesson} - {page}.")
```

```python
for row in df_all_pages.iterrows():
  url_id = row[1]['page_id']
  lesson = row[1]['page_lesson']
  page = row[1]['page_number']

  save_html(url_id, lesson, page)
```
